## Recupera os dados clusterizados anteriormente, treina o modelo, fazendo a classificação da qualidade do ar

```python
# importa bibliotecas
import pandas as pd
from sklearn import metrics # algoritmo de métricas de desempenho
from sklearn.linear_model import LogisticRegression # algoritmo de modelo de regressão logística
from sklearn.model_selection import train_test_split # algoritmo de separação de dados de treino e teste

pd.set_option('display.max_rows', None) # permite exibição de todas as linhas
pd.set_option('display.max_columns', None) # permite exibição de todas as colunas
import warnings
warnings.simplefilter('ignore') # desativa avisos de warning
```

```python
# carregando arquivo tratado em dataframe
df_quality_clust = pd.read_csv('arquivos_tratados/df_quality_clustered.csv')
```

```python
# conferindo a dimensão do dataframe, linhas e colunas
df_quality_clust.shape
```

```python
# ordenando pelo index o dataframe carregado
df_quality_clust = df_quality_clust.sort_index()
```

```python
# reconhecendo as primeiras linhas do df
df_quality_clust.head(5)
```

```python
list(df_quality_clust)
```

```python
list_og = [
 'CO/hr',
 'C6H6-micgr/m3/hr',
 'NMHC/hr',
 'NOx-ppb/hr',
 'NOx/hr',
 'NO2-micgr/m3/hr',
 'NO2/hr',
 'O3/hr',
 'temperatura°C',
 'umidade_relativa',
 'umidade_absoluta',
 'dia_semana',
 'hour',
 'cluster',
]

list_trans = [
 'CO/hr',
 'C6H6-micgr/m3/hr',
 'NMHC/hr',
 'NOx-ppb/hr',
 'NOx/hr',
 'NO2-micgr/m3/hr',
 'NO2/hr',
 'O3/hr',
 'temperatura°C',
 'umidade_relativa',
 'umidade_absoluta',
 'one-hot__dia_semana_0.0',
 'one-hot__dia_semana_1.0',
 'one-hot__dia_semana_2.0',
 'one-hot__dia_semana_3.0',
 'one-hot__dia_semana_4.0',
 'one-hot__dia_semana_5.0',
 'one-hot__dia_semana_6.0',
 'one-hot__hour_0.0',
 'one-hot__hour_1.0',
 'one-hot__hour_2.0',
 'one-hot__hour_4.0',
 'one-hot__hour_5.0',
 'one-hot__hour_6.0',
 'one-hot__hour_7.0',
 'one-hot__hour_8.0',
 'one-hot__hour_9.0',
 'one-hot__hour_10.0',
 'one-hot__hour_11.0',
 'one-hot__hour_12.0',
 'one-hot__hour_13.0',
 'one-hot__hour_14.0',
 'one-hot__hour_15.0',
 'one-hot__hour_16.0',
 'one-hot__hour_17.0',
 'one-hot__hour_18.0',
 'one-hot__hour_19.0',
 'one-hot__hour_20.0',
 'one-hot__hour_21.0',
 'one-hot__hour_22.0',
 'one-hot__hour_23.0',
 'cluster',
]
```

```python
df_quality_trans = df_quality_clust[list_trans]
df_quality_og = df_quality_clust[list_og]
```

```python
df_quality_trans.shape
```

```python
df_quality_trans.head(5)
```

```python
# separa variáveis independentes(x) e dependentes(y)
X = df_quality_trans.drop(columns=['cluster'], axis=1)
y = df_quality_trans[['cluster']]
```

```python
# separa os sets de treino e teste em 20% para teste e 80% para treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# verifica o tamanho de cada set resultante do split
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# treina o modelo de classificação
clf = LogisticRegression(random_state=42)
clf.fit(X_train, y_train)
```

```python
# prediz y com predict e predict_proba gerando o percentual de cada classe
y_pred_proba = clf.predict_proba(X_test) # fornece o percentual de cada classe
y_pred = clf.predict(X_test)
```

```python
# visualizando a predição de y de X_test
y_pred
```

```python
# gera a matriz de confusão para análise
# soma da diagonal principal sobre a soma total é o percentual de acerto (acurácia, precisão, recall)
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
# verifica a acurácia, precisão e recall do resultado que corresponde à matriz de confusão
print('acurácia =', metrics.accuracy_score(y_test, y_pred))
print('precisão =', metrics.precision_score(y_test, y_pred, average='weighted'))
print('recall   =', metrics.recall_score(y_test, y_pred, average='weighted'))
print('f1-score =', metrics.f1_score(y_test, y_pred, average='weighted'))
```

```python
# visualiza a predição das probabilidades de y
print(y_pred_proba)
```

```python
# transforma array em dataframe
dfproba = pd.DataFrame(y_pred_proba)
```

```python
# renomeia as colunas de resultado do predict_proba
dfproba.columns = ['Ruim', 'Media', 'Boa']
```

```python
# arredonda os resultados com duas casas decimai
dfproba = round(dfproba*100, 2)
```

```python
# soma o total de percentuais para checar se fecha em 100%
dfproba['soma_perc'] = dfproba['Ruim'] + dfproba['Media'] + dfproba['Boa']
```
